```python
#using https://newsapi.org/
###Cons: 1.Can request only 100 times 2.Can search up to only a month old articles
def news_collect(keyword, date, api_keys):
    import requests
    import json
    url = ('https://newsapi.org/v2/everything?' + 'q='+ keyword + '&from=' + date + '&sortBy=popularity' + '&apiKey=' + api_keys)
    response = requests.get(url)
    news = response.json()
    total_results = news['totalResults']
    if total_results > 0:
        print(f"Total Results = {total_results}")
        news_data = pd.DataFrame(news['articles'])
        sources = news_data['source'].apply(lambda x : x['name']).tolist()
        news_data['source'] = sources
        return news_data
    else:
        print('No Results Found')
```

```python
def news_collector(access_key, date, sort_by, keywords, category):
    import http.client, urllib.parse
    import yaml

    conn = http.client.HTTPConnection('api.mediastack.com')

    params = urllib.parse.urlencode({
    'access_key': access_key,
    'date': date,
    'sort': sort_by,
    'keywords': keywords,
    'category': category
    })

    conn.request('GET', '/v1/news?{}'.format(params))

    res = conn.getresponse()
    data = res.read()

    result = yaml.safe_load(data.decode('utf-8'))
    
    total_results = result['pagination']['total']
    
    if total_results > 0:
        print(f"Total Results = {total_results}")
        news_data = pd.DataFrame(result['data'])
        return news_data
    else:
        print('No Results Found')
```

```python
news_collector('6422f292cb44f760be067f73f8d5a3d7', '2023-02-01', 'popularity', 'ocean technology', 'climate')
```
