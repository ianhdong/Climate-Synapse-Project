```python
def gov_collect(collection, date, offset, pageSize, api_key):
    url = "https://api.govinfo.gov/collections/" + collection + "/%20" + date[:13] + "%3A" + date[14:16] + "%3A" +  date[17:20] + "?offset=" + str(offset) + "&pageSize=" + str(pageSize) + "&api_key=" + api_key
    response = requests.get(url)
    gov_data = response.json()
    total_results = gov_data['count']
    if total_results > 0:
        print(f"Total Results = {total_results}")
        gov_data_pd = pd.DataFrame(gov_data['packages'])
        return gov_data_pd
    else:
        print('No Results Found')
```

```python
def transportation_fuels(api_key):
    url = "https://developer.nrel.gov/api/vehicles/v1/fuels.json?api_key=" + api_key
    response = requests.get(url)
    transportation_data = response.json()
    total_results = transportation_data['metadata']['resultset']['count']
    if total_results > 0:
        print(f"Total Results = {total_results}")
        transportation_data_pd = pd.DataFrame(transportation_data['result'])
        return transportation_data_pd
    else:
        print('No Results Found')
```

```python
def transportation_incentives_laws(api_key, limit=100, jurisdiction="", technology="", incentive_type="", regulation_type="", user_type="", poc=False, recent=False, expired=False, law_type="", key_word="", local=False):
    url = "https://developer.nrel.gov/api/transportation-incentives-laws/v1.json?api_key=" + api_key + "&limit=" + str(limit) + "&jurisdiction=" + jurisdiction + "&technology=" + technology + "&incentive_type=" + incentive_type + "&regulation_type=" + regulation_type + "&user_type=" + user_type + "&poc=" + str(poc) + "&recent=" + str(recent) + "&expired=" + str(expired) + "&law_type=" + law_type + "&keyword=" + key_word + "&local=" + str(local)
    response = requests.get(url)
    transportation_laws_categories_data = response.json()
    total_results = transportation_laws_categories_data['metadata']['count']
    if total_results > 0:
        print(f"Total Results = {total_results}")
        transportation_laws_categories_data_pd = pd.DataFrame(transportation_laws_categories_data['result'])
        return transportation_laws_categories_data_pd
    else:
        print('No Results Found')
```
